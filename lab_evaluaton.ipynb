```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

#Load the data
ml = pd.read_excel("Online_Retail.xlsx")
```

```python
#display 10 rows
print(ml.head(10))
```

```python
print(ml.columns)
```

```python
#missing values
print(ml.isnull)
```

```python
#check for rows and missing values
print(ml.isnull().sum())
```

```python
#remove the duplicate values
ml.drop_duplicates()
```

```python
#add a column revenue = quantity * unit price
ml["Revenue"] = ml["Quantity"] * ml["UnitPrice"]
```

```python
#find top 10 selling products based on quantity
top_products = ml.groupby("Description")["Quantity"].sum().sort_values(ascending=False).head(10)
print(top_products)
```

```python
#calculate the total revenue of the store
total_revenue = ml["Revenue"].sum()
print(total_revenue)
```

```python
#total no. of unique customers
total_customers = ml["CustomerID"].unique()
print(total_customers)
```

```python
#calc the avg oder value
avg_order_value = ml["Revenue"].mean()
print(avg_order_value)
```

```python
#identify the country with the highest no. of orders
top_country = ml["Country"].value_counts().idxmax()
print(top_country)
```

```python
#list all the products which are brought together
products_together = ml.groupby("InvoiceNo")["Description"].apply(list)
print(products_together)
```

```python
#identify the customer who purchased more than 10 items
customers = ml[ml["Quantity"] > 10]["CustomerID"]
print(customers)
```

```python
#find the peak shopping hours of the customers
ml["InvoiceDate"] = pd.to_datetime(ml["InvoiceDate"])
ml["Hour"] = ml["InvoiceDate"].dt.hour
peak_hours = ml["Hour"].value_counts()
print(peak_hours)
```
